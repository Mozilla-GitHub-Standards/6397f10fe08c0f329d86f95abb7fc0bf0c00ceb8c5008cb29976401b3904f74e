```python
import numpy as np
import pandas as pd
import os, os.path
```

```python
def create_empty_df(n,col_names):
    df = pd.DataFrame('', index=range(n), columns=col_names)
    return df
```

## Data Integration

```python
target_folder_path = 'Data/2017_10_16/'
col_names = ['Date','Version','Emotion','Rating','Reviews','Store']
```

```python
import glob
# Read in all the dataset
file_paths = glob.glob(target_folder_path + '*')
for file_path in file_paths:
    file_format = file_path.split('.')[-1]
    if file_format == 'xlsx':
        #df = pd.ExcelFile(file_path)
        #df = pd.read_excel(open(file_path,'rb'), sheetname='Sheet 1')
        xl = pd.ExcelFile(file_path)
        SurveyGizmo = xl.parse(xl.sheet_names[0]).fillna('')
    else:
        Appbot = pd.read_csv(file_path).fillna('')
```

```python
Appbot.head()
```

```python
target_col_Appbot = ['Store','Version','Rating','Date','Subject','Body','Translated Subject','Translated Body','Emotion']
```

```python
Appbot_Processed = create_empty_df(len(Appbot),col_names)
```

```python
Appbot_Processed['Date'] =  Appbot['Date']
Appbot_Processed['Version'] =  Appbot['Version']
Appbot_Processed['Rating'] =  Appbot['Rating']
Appbot_Processed['Store'] =  Appbot['Store']
Appbot_Processed['Emotion'] =  Appbot['Emotion']
Appbot_Processed['Reviews'] = Appbot[['Subject','Body']].apply(lambda x : '{}. {}'.format(x[0],x[1]), axis=1)
Appbot_Processed['Translated Reviews'] = Appbot[['Translated Subject','Translated Body']].apply(lambda x : '{}. {}'.format(x[0],x[1]), axis=1)
```
